# Data Cleaning

**Renaming certain columns in 2017 dataset**

```python
a_2017.rename(columns={"Happiness.Rank" : "Happiness Rank", "Happiness.Score" : "Happiness Score", "Economy..GDP.per.Capita.":"Economy (GDP per Capita)", "Trust..Government.Corruption.":"Trust (Government Corruption)", "Dystopia.Residual":"Dystopia Residual", "Health..Life.Expectancy." : "Health (Life Expectancy)"}, inplace= True)
a_2018.rename(columns={"Score":"Happiness Score", "Country or region":"Country","Freedom to make life choices":"Freedom", "Healthy life expectancy":"Health (Life Expectancy)", "GDP per capita":"Economy (GDP per Capita)", "Social support":"Family", "Perceptions of corruption":"Trust (Government Corruption)"}, inplace= True)
a_2019.rename(columns={"Score":"Happiness Score", "Country or region":"Country","Freedom to make life choices":"Freedom", "Healthy life expectancy":"Health (Life Expectancy)", "GDP per capita":"Economy (GDP per Capita)", "Social support":"Family", "Perceptions of corruption":"Trust (Government Corruption)"}, inplace= True)
```

**Setting Happiness Rank / Overall Rank as the index**

```python
a_2015.set_index("Happiness Rank",inplace=True)
a_2016.set_index("Happiness Rank",inplace=True)
a_2017.set_index("Happiness Rank",inplace=True)
a_2018.set_index("Overall rank",inplace=True)
a_2019.set_index("Overall rank",inplace=True)
```

**Dropping few irrelavant (to our analysis) columns from 2015-2017 datasets so that they are similar to 2018 & 2019 datasets**

```python
drop15 = ["Standard Error","Region"]
drop16 = ["Lower Confidence Interval", "Upper Confidence Interval","Region"]
drop17 = ["Whisker.high", "Whisker.low"]
```

```python
a_2015.drop(columns= drop15, inplace=True)
a_2016.drop(columns= drop16, inplace=True)
a_2017.drop(columns= drop17, inplace=True)
```

**Checking for null values**

```python
mask15= a_2015.isnull().sum()
mask16= a_2016.isnull().sum()
mask17= a_2017.isnull().sum()
mask18= a_2018.isnull().sum()
mask19= a_2019.isnull().sum()
```

```python
print("2015","\n",mask15,"\n\n","2016","\n", mask16,"\n\n","2017","\n", mask17,"\n\n","2018","\n", mask18,"\n\n","2019","\n", mask19)
```

**There is one null value in  2018 data which we will fill by the mean value**

```python
a_2018["Trust (Government Corruption)"]= a_2018["Trust (Government Corruption)"].fillna(a_2018["Trust (Government Corruption)"].mean())
```

# 1. Happiness Score of top 25 Countries (2015-2019 : Mean value considered)

```python
combined= pd.concat([a_2015.head(25), a_2016.head(25), a_2017.head(25), a_2018.head(25), a_2019.head(25)])   #Concatenating the top 25 data from each dataset

m_com= combined.groupby("Country")["Happiness Score"].mean()
df= pd.DataFrame(data= m_com)
df= df.reset_index()
df.sort_values(by= "Happiness Score", ascending = False, inplace= True)
df

#we combined the data for top 25 countries from each dataset and found the mean for Happiness score. A dataframe is then created which gives us a better view at the happiness score of top 25 countries.
```

```python
plt.figure(figsize=(20,8))
plt.bar("Country", "Happiness Score", data=df.head(25))
plt.xticks(rotation=90)
plt.title("Happiness Score of top 25 Countries (2015-2019)")
plt.xlabel("Country")
plt.ylabel("Happiness Score")
plt.show()
```

Findings in (1):
* We got the data for 32 countries and after considering the mean of happiness score, we took top 25 values of the sorted(highest to lowest) data from year 2015-2019
* We plotted a bar chart of top 25 countries showcasing their mean happiness score
* Denmark has the highest mean value (2015-2019) in the top 25 data
* Singapore has the lowest mean value (2015-2019) in the top 25 data

# 2. Happiness Score of bottom 25 Countries (2015-2019 : Mean taken)

```python
combineds= pd.concat([a_2015.tail(25), a_2016.tail(25), a_2017.tail(25), a_2018.tail(25), a_2019.tail(25)])       #Concatenating the last 25 data from each dataset

m_coms= combineds.groupby("Country")["Happiness Score"].mean()
dfs= pd.DataFrame(data= m_coms)
dfs= dfs.reset_index()
dfs.sort_values(by="Happiness Score", inplace= True)
dfs

#we combined the data for last 25 countries from each dataset and found the mean for Happiness score. A dataframe is then created which gives us a better view at the happiness score of top 25 countries.
```

```python
plt.figure(figsize=(20,8))
plt.bar("Country", "Happiness Score", data=dfs.head(25))
plt.xticks(rotation=90)
plt.title("Happiness Score of bottom 25 Countries (2015-2019)")
plt.xlabel("Country")
plt.ylabel("Happiness Score")
plt.show()
```

Findings in (2):
* We got the data for 39 countries and after considering the mean of happiness score, we took top 25 values of the sorted(lowest-highest) data from year 2015-2019
* We plotted a bar chart of last 25 countries showcasing their mean happiness score
* Comoros has the highest mean value (2015-2019) in the bottom 25 data
* Burundi has the lowest mean value (2015-2019) in the bottom 25 data

# 3. GDP of top 5 countries (2015-2019)

```python
combined5= pd.concat([a_2015.head(), a_2016.head(), a_2017.head(), a_2018.head(), a_2019.head()])
combined5

#Concatenating the top 5 data from each dataset
```

```python
combined5.set_index(["Year","Country"])
```

```python
uni= combined5["Country"].unique()
uni

#finding unique values of the countries found after concatenating
```

```python
c1 = combined5[combined5['Country'] == 'Switzerland']
c2 = combined5[combined5['Country'] == 'Iceland']
c3 = combined5[combined5['Country'] == 'Denmark']
c4 = combined5[combined5['Country'] == 'Norway']
c5 = combined5[combined5['Country'] == 'Canada']
c6 = combined5[combined5['Country'] == 'Finland']
c7 = combined5[combined5['Country'] == 'Netherlands']

#seperating data for sepearte countries for the 5 years
```

```python
plt.figure(figsize=(16,8))

plt.plot(c1['Year'], c1['Economy (GDP per Capita)'], label='Switzerland',marker='o')
plt.plot(c2['Year'], c2['Economy (GDP per Capita)'], label='Iceland',marker='o')
plt.plot(c3['Year'], c3['Economy (GDP per Capita)'],label='Denmark',marker='o')
plt.plot(c4['Year'], c4['Economy (GDP per Capita)'], label='Norway',marker='o')
plt.plot(c5['Year'], c5['Economy (GDP per Capita)'], label='Canada',marker='o')
plt.plot(c6['Year'], c6['Economy (GDP per Capita)'], label='Finland',marker='o')
plt.plot(c7['Year'], c7['Economy (GDP per Capita)'],label='Netherlands',marker='o')

plt.title('Line Trend Chart showing GDP per Capita for top 5 countries(2015-2019)')
plt.xlabel('Year')
plt.ylabel('Economy (GDP per Capita)')
plt.legend()
plt.show()
```

Findings in (3)
* We got data for 7 countries while considering top 5 countries wrt GDP per capita
* We plotted a line chart combining the data of each country
* The data for Netherlands and Canada is not available on line plot because they were in the top 5 only once in 5 years so there is no enough data
* Finland is missing 2015 data since it was not in the top 5 in 2015. Similarly, Switzerland is missing 2019 data since it was not in the top 5 in 2019
